```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
data = pd.read_csv('IRIS.csv')

# Install required packages
import sys
import subprocess
try:
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', '--user', 'ydata-profiling'])
except subprocess.CalledProcessError as e:
    print(f"Error installing package: {e}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from ydata_profiling import ProfileReport

# Generate profile report using the already loaded 'data' DataFrame
profile = ProfileReport(data, title="Iris Dataset EDA Report", explorative=True)

# Save the report to HTML file
profile.to_file("EDA_Iris.html")

print("EDA report saved as EDA_Iris.html")
```
